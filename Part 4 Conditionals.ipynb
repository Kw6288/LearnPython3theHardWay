# 习题32 循环和列表

```python
the_count =  [1,2,3,4,5]
fruits = ['apples','organges','pears','apricots']
change = [1,'pennies',2,'dimes',3,'quarters']

# this first kind of for-loop goes through a list

for number in the_count:
    print(f"This is coount {number}")
    
for fruit in fruits:
    print(f"A fruit of type: {fruit}")

# also we can go through mixed lists too
# notice we have to use {} since we dont know what's in it

for i in change:
    print(f"I got {i}")

# we can also build lists, first start with an empty one

elements = []

# then use the range function to do 0 to 5 counts
for i in range(0,6):
    print(f"Adding {i} to the list")
    
    # append is a function that lists understand
    elements.append(i)

# now we can print them out too
for i in elements:
    print(f"elements was:{i}")
    
```

# 习题33 while循环

```python
i = 0
numbers = []

while i < 6:
    print(f"At the topi i is {i}") #打印出第一个i 的数值
    numbers.append(i)
    
    i = i+1
    print("number now: ", numbers)
    print(f"At the bottom i is {i}") #打印出最后一个 i 的数值
    
print("The numbers: ")
    
for num in numbers:
    print(num)
        
```

```python
# 巩固练习1.



def list_generator(x,y):
    i = 0
    numbers = []
    while i < x:
        print(f"At the topi i is {i}") #打印出第一个i 的数值
        numbers.append(i)

        i = i + y
        print("number now: ", numbers)
        print(f"At the bottom i is {i}") #打印出最后一个 i 的数值

list_generator(100,3)
```
